```python
import numpy as np
import matplotlib.pyplot as plt

# 예제 데이터 생성
np.random.seed(42)
X = np.random.rand(100, 2) * 10  # 2차원 데이터를 0부터 10 사이의 랜덤한 값으로 생성
y = 2 * X[:, 0] + 3 * X[:, 1] + np.random.randn(100)  # y = 2x1 + 3x2 + 잡음

# 데이터 시각화
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Target')
plt.title('Original Data')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, Normalizer

# StandardScaler를 사용하여 데이터 표준화
scaler_standard = StandardScaler()
X_standard = scaler_standard.fit_transform(X)

# MinMaxScaler를 사용하여 최소-최대 정규화
scaler_minmax = MinMaxScaler()
X_minmax = scaler_minmax.fit_transform(X)

# RobustScaler를 사용하여 중앙값과 IQR을 이용한 스케일링
scaler_robust = RobustScaler()
X_robust = scaler_robust.fit_transform(X)

# Normalizer를 사용하여 데이터 정규화
scaler_normalizer = Normalizer()
X_normalized = scaler_normalizer.fit_transform(X)

# 전처리 적용 후 시각화
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.scatter(X_standard[:, 0], X_standard[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.title('StandardScaler')

plt.subplot(2, 2, 2)
plt.scatter(X_minmax[:, 0], X_minmax[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1 (MinMax Scaled)')
plt.ylabel('Feature 2 (MinMax Scaled)')
plt.title('MinMaxScaler')

plt.subplot(2, 2, 3)
plt.scatter(X_robust[:, 0], X_robust[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1 (Robust Scaled)')
plt.ylabel('Feature 2 (Robust Scaled)')
plt.title('RobustScaler')

plt.subplot(2, 2, 4)
plt.scatter(X_normalized[:, 0], X_normalized[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1 (Normalized)')
plt.ylabel('Feature 2 (Normalized)')
plt.title('Normalizer')

plt.tight_layout()
plt.show()
```
