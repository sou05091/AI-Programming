```python
## 훈련 세트, 테스트 세트를 나눕니다.
from sklearn.model_selection import train_test_split
import mglearn
X,y = mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# 이웃의 수를 3으로 설정합니다.
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
# 훈련 세트를 사용하여 분류 모델을 학습시킵니다.
clf.fit(X_train, y_train)
```

```python
# predict 메서드를 호출하여 예측합니다.
print("테스트 세트 예측", clf.predict(X_test))
```

```python
# 모델이 얼마나 잘 일반화 하였는지 평가하기 위해 score 메서드에 테스트 데이터와 테스트 레이블을 넣어 호출합니다.
print("테스트 세트 정확도: {:.2f}".format(clf.score(X_test, y_test)))
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1,3, figsize =(10,3))
for n_neighbors,ax in zip([1,3,9],axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    mglearn.plots.plot_2d_separator(clf, X, file=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    ax.set_title("{}이웃".format(n_neighbors))
    ax.set_xlabel("특성0")
    ax.set_ylabel("특성1")
axes[0].legend(loc=3)
    
```

```python
def isTripleConsecutive(word):
    for i in range(len(word) - 2):
        letter1 = word[i]
        letter2 = word[i + 1]
        letter3 = word[i + 2]
        if ord(letter2) == ord(letter1) + 1 and ord(letter3) == ord(letter2) + 1:
            return True
    return False

# Prompt the user to enter a word
word = input("Enter a word: ")

# Check if the word contains three consecutive letters in consecutive alphabetical order
if isTripleConsecutive(word.upper()):
    print(f"{word} contains three successive letters in consecutive alphabetical order.")
else:
    print(f"{word} does not contain three successive letters in consecutive alphabetical order.")
```

```python
# w는 lr객체의 coef_속성에 저장되었고, b는 intercept_속성에 저장되었습니다.
from sklearn.linear_model import LinearRegression
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
lr = LinearRegression().fit(X_train, y_train)
```

```python
# 훈련 세트와 테스트 세트의 점수가 매우 비슷하다는 것을 알 수 있습니다.
# 이는 과소적합인 상태이라는 점을 알 수 있습니다.
# 1차원 데이터셋에서는 모델이 매우 단순하므로 과데적합을 걱정할 필요는 없습니다.
print("훈련 세트 점수:{:.2f}".format(lr.score(X_train, y_train)))
print("테스트 세트 점수:{:.2f}".format(lr.score(X_test, y_test)))
```

```python
# 이전과 같은 방식으로 복잡한 데이터셋인 보스턴 주택가격에 대해서는 어떻게 동작하는지 보겠습니다. 
X, y= mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
lr = LinearRegression().fit(X_train, y_train)

print("훈련 세트 점수:{:.2f}".format(lr.score(X_train, y_train)))
print("테스트 세트 점수:{:.2f}".format(lr.score(X_test, y_test)))
```
